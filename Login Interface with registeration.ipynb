```python
#import the modules required

from tkinter import *
from PIL import ImageTk
from tkinter import messagebox
import pymysql

class Login:
    def __init__(self,root):
        self.root=root
        self.root.title("Login and Registeration system")
        self.root.geometry("1800x1200")
        self.loginform()
        
    def loginform(self):
        Frame_login=Frame(self.root,bg='#B0E0E6')
        Frame_login.place(x=0,y=0,height=1200,width=1800)
        
        self.img = ImageTk.PhotoImage(file="GDPicon.jpg")
        img = Label(Frame_login,image=self.img).place(x=670,y=100,width=600,height=560)
        
        frame_input=Frame(self.root, bg='white')
        frame_input.place(x=320,y=100,height=560,width=350)
        
        label1=Label(frame_input,text="Login Here",font=('impact',32,'bold'),fg="black",bg='white')
        label1.place(x=75,y=50)
        
        label2=Label(frame_input,text="Email Id",font=('Arial',20,'bold'),fg="#B22222",bg='white')
        label2.place(x=30,y=135)
        
        self.email_txt=Entry(frame_input,font=("times new roman",15,"bold"),bg='lightgray')
        self.email_txt.place(x=30,y=175,width=270,height=35)
        
        label3=Label(frame_input,text="Password",font=('Arial',20,'bold'),fg="#B22222",bg='white')
        label3.place(x=30,y=235)
        
        self.password=Entry(frame_input,font=("Times new roman",15,"bold"),bg='lightgray')
        self.password.place(x=30,y=275,width=270,height=35)
        
        btn1=Button(frame_input,text="Forgot Password?",cursor='hand2',command=lambda:self.fwindow(),font=('calibri',14),bg='white',fg='black',bd=0)
        btn1.place(x=75,y=355)
        
        btn2=Button(frame_input,command=self.login,text="LOGIN",cursor='hand2',font=('Times New Roman',15),bg='#B22222',fg='white',bd=0,width=15,height=1)
        btn2.place(x=65,y=390)
        
        btn3=Button(frame_input,text="Not Registered Yet? \n Register Now",command=self.Register,cursor="hand2",font=('calibri',14),bg='white',fg="black",bd=0)
        btn3.place(x=70,y=435)
        
    def login(self):
        if self.email_txt.get()=="" or self.password.get()=="":
            messagebox.showerror("Error","All fields are required",parent=self.root)
        else:
            try:
                con=pymysql.connect(host='localhost',user='root',password='nida3694',database='registerationdb')
                cur=con.cursor()
                cur.execute('Select * from register where emailid=%s and password=%s',(self.email_txt.get(),self.password.get()))
                row=cur.fetchone()
                if row==None:
                    messagebox.showerror('Error','Invalid Username and Password',parent=self.root)
                    self.loginclear()
                    self.email_txt.focus()
                else:
                    self.appscreen()
                    con.close()
            except Exception as es:
                messagebox.showerror('Error',f'Error Due to : {str(es)}',parent=self.root)
                
    def Register(self):
        Frame_login=Frame(self.root,bg="#FFDAB9")
        Frame_login.place(x=0,y=0,height=1200,width=1800)
        
        self.img = ImageTk.PhotoImage(file="Registerimg.jpg")
        img = Label(Frame_login,image=self.img).place(x=820,y=100,width=612,height=612)
        
        frame_input2=Frame(self.root, bg='white')
        frame_input2.place(x=190,y=100,height=612,width=630)
        
        label1=Label(frame_input2,text="REGISTER HERE",font=('impact',34,'bold'),fg="black",bg='white')
        label1.place(x=30,y=20)
        
        label2=Label(frame_input2,text="Username",font=("Comic Sans MS",20,"bold"),fg='blue',bg='white')
        label2.place(x=30,y=95)

        self.entry=Entry(frame_input2,font=("times new roman",15,"bold"), bg='lightgray')
        self.entry.place(x=30,y=145,width=270,height=35)
        
        label3=Label(frame_input2,text="Password",font=("Comic Sans MS",20,"bold"),fg='blue',bg='white')
        label3.place(x=30,y=195)
        
        self.entry2=Entry(frame_input2,font=("times new roman",15,"bold"),bg='lightgray')
        self.entry2.place(x=30,y=245,width=270,height=35)
        
        label4=Label(frame_input2,text="Email-id",font=("Comic Sans MS",20,"bold"),fg='blue',bg='white')
        label4.place(x=330,y=95)
        
        self.entry3=Entry(frame_input2,font=("times new roman",15,"bold"),bg='lightgray')
        self.entry3.place(x=330,y=145,width=270,height=35)

        label5=Label(frame_input2,text="Confirm Password",font=("Comic Sans MS",20,"bold"),fg='blue',bg='white')
        label5.place(x=330,y=195)

        self.entry4=Entry(frame_input2,font=("times new roman",15,"bold"),bg='lightgray')
        self.entry4.place(x=330,y=245,width=270,height=35)
        
        btn2=Button(frame_input2,command=self.register,text="REGISTER",cursor="hand2",font=("times new roman",17),fg="white",bg="purple",bd=0,width=15,height=2)
        btn2.place(x=200,y=340)
        
        btn3=Button(frame_input2,command=self.loginform,text="Already Registered? \n Login",cursor="hand2",font=("calibri",18),bg='white',fg="black",bd=0)
        btn3.place(x=200,y=450)
        
    def register(self):
        if self.entry.get()==""or self.entry2.get()==""or self.entry3.get()==""or self.entry4.get()=="":
                messagebox.showerror("Error","All Fields Are Required",parent=self.root)
        elif self.entry2.get()!=self.entry4.get():
                messagebox.showerror("Error","Password and Confirm Password Should Be Same",parent=self.root)
        else:
            try:
                con=pymysql.connect(host="localhost",user="root",password="nida3694",database="registerationdb")
                cur=con.cursor()
                cur.execute("select * from register where emailid=%s",self.entry3.get())
                row=cur.fetchone()

                if row!=None:
                    messagebox.showerror("Error","User already Exist,Please try with another Email",parent=self.root)
                    self.regclear()
                    self.entry.focus()
                    
                else:
                    cur.execute("insert into register values(%s,%s,%s,%s)",(self.entry.get(),self.entry3.get(),self.entry2.get(),self.entry4.get()))
                    con.commit()
                    con.close()
                    messagebox.showinfo("Success","Register Succesfull",parent=self.root)
                    self.regclear()

            except Exception as es:
                messagebox.showerror("Error",f"Error due to:{str(es)}",parent=self.root)

    def appscreen(self):
        Frame_login=Frame(self.root,bg="orange")
        Frame_login.place(x=0,y=0,height=700,width=1366)
        label1=Label(Frame_login,text="Hi! Welcome GDP Prediction",font=('times new roman',32,'bold'),fg="black",bg='white')
        label1.place(x=375,y=100)
        btn2=Button(Frame_login,text="Logout",command=self.loginform,cursor="hand2",font=("times new roman",15),fg="white",bg="orangered",bd=0,width=15,height=1)
        btn2.place(x=1000,y=10)

    def regclear(self):
        self.entry.delete(0,END)
        self.entry2.delete(0,END)
        self.entry3.delete(0,END)
        self.entry4.delete(0,END)

    def loginclear(self):
        self.email_txt.delete(0,END)
        self.password.delete(0,END)
        
    def fwindow(self):
        forgetwindow=Toplevel()
        
        self.forgetwindow=forgetwindow
        self.forgetwindow.title("Forget Password")
        self.forgetwindow.geometry("800x600")
        self.forgetwindow.resizable(False,False)
        
        self.username=Label(self.forgetwindow,text="Email Id",font=("Times new roman",18,'bold'))
        self.username.place(x=100,y=90)
        
        self.entry5=Entry(self.forgetwindow,font=("times new roman",16))
        self.entry5.place(x=100,y=140,width=250)
        
        self.newpass=Label(self.forgetwindow,text="New Password",font=("Times new roman",18,'bold'))
        self.newpass.place(x=100,y=198)
        
        self.entry6=Entry(self.forgetwindow,font=("times new roman",16))
        self.entry6.place(x=100,y=240,width=250)
        
        self.conpass=Label(self.forgetwindow,text="Confirm Password",font=("Times new roman",18,'bold'))
        self.conpass.place(x=100,y=290)
        
        self.entry7=Entry(self.forgetwindow,font=("times new roman",16))
        self.entry7.place(x=100,y=340,width=250)
        
        self.chbutton = Button(self.forgetwindow,text='Change password',font=("times new roman",15,'bold'),bd=5,activebackground="lightgreen",command=lambda:self.chpassword())
        self.chbutton.place(x=100,y=390,width=250)
        
    
    def chpassword(self):
        if self.entry5.get=='' and self.entry6.get=='' and self.entry7.get=='':
            messagebox.showerror("Warning ,All fields are required")
            
        else:
            con=pymysql.connect(host="localhost",user="root",password="nida3694",database="registerationdb")
            cur=con.cursor()
            cur.execute("Update register set password=%s where username=%s ",(self.entry5.get(),self.entry6.get()))
            con.commit()
            con.close()
            messagebox.showinfo("Successfully Password Changed")
        
root=Tk()
ob=Login(root)
root.mainloop()
        
        
```
